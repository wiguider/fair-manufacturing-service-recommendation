```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.utils.seed import set_seed
from src.utils.config import load_config
from src.data.mskg_processor import process_mskg
from src.data.supply_chain_processor import process_dataco

set_seed(42)
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Load MSKG dataset
mskg = process_mskg(seed=42)
print(mskg.summary())
```

```python
# Size group distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
mskg.protected_attrs['size_group'].value_counts().plot.bar(ax=axes[0], title='Size Distribution')
mskg.protected_attrs['geo_group'].value_counts().plot.bar(ax=axes[1], title='Geographic Distribution')
plt.tight_layout()
plt.show()
```

```python
# Popularity bias analysis
item_pop = mskg.interactions.groupby('item_id').size().reset_index(name='count')
merged = item_pop.merge(mskg.protected_attrs[['item_id', 'size_group']], on='item_id')

fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=merged, x='size_group', y='count', order=['small', 'medium', 'large'], ax=ax)
ax.set_title('Interaction count by manufacturer size')
ax.set_ylabel('Number of interactions')
plt.show()

print('Mean interactions by size group:')
print(merged.groupby('size_group')['count'].mean())
```
